```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
north_america = pd.read_csv('north_america_2000_2010.csv', index_col=0)
```

```python
south_america = pd.read_csv('south_america_2000_2010.csv', index_col=0)
```

```python
north_america
```

```python
south_america
```

```python
sns.lineplot(data=north_america)
```

```python
north_america.T.plot(title="Average Labor Hours Per Year")
```

```python
south_america.T.plot(title="Average Labor Hours Per Year")
```

```python
americas = pd.concat([north_america,south_america])
```

```python
americas
```

```python
america_1 = pd.read_csv('americas_2011.csv', index_col=0)
america_2 = pd.read_csv('americas_2012.csv', index_col=0)
america_3 = pd.read_csv('americas_2013.csv', index_col=0)
america_4 = pd.read_csv('americas_2014.csv', index_col=0)
america_5 = pd.read_csv('americas_2015.csv', index_col=0)
```

```python
temp=america_1.join(america_2)
```

```python
temp=temp.join(america_3)
```

```python
temp=temp.join(america_4)
```

```python
temp=temp.join(america_5)
```

```python
temp
```

```python
americas = americas.join(temp)
```

```python
americas.index.names = ['Country']
```

```python
americas
```

```python
americas.T.plot(title="Average Labor Hours Per Year")
```

```python
asia = pd.read_csv('asia_2000_2015.csv', index_col=0)
asia
```

```python
europe = pd.read_csv('europe_2000_2015.csv', index_col=0)
europe.head()
```

```python
south_pacific = pd.read_csv('south_pacific_2000_2015.csv', index_col=0)
south_pacific
```

```python
world = pd.concat([americas,europe,south_pacific,asia])
world
```

```python
world.index
```

```python
world.T.plot(title="Average Labor Hours Per Year")
```
